# Bhojpuri Text Normalization Test Cases

This notebook tests the Bhojpuri text normalization implementation.

```python
import sys
import os

# Add the NeMo-text-processing directory to the path
sys.path.insert(0, os.path.abspath('.'))

from nemo_text_processing.text_normalization.normalize import Normalizer

print("Imports successful!")
```

```python
# Initialize Bhojpuri normalizer
normalizer_bho = Normalizer(
    input_case='cased',
    lang='bho',
    cache_dir=None,  # Set to a directory path if you want to cache .far files
    overwrite_cache=False,
    post_process=True
)

print("Bhojpuri Text Normalizer initialized successfully!")
print(f"Language: {normalizer_bho.lang}")
```

## 1. Cardinal Numbers Test



```python
cardinal_tests = [
    "155",
    "१२३४",
    "१२३४५",
    "१२३४५६",
    "१२३४५६७",
    "-१२३",
    "-120",
    "१०००",
    "१०००००",  # 1 lakh
    "१०००००००",  # 1 crore
]

print("=" * 60)
print("CARDINAL NUMBERS TEST")
print("=" * 60)
for test in cardinal_tests:
    result = normalizer_bho.normalize(test)
    print(f"Input:  {test:15} -> Output: {result}")
```

## 2. Decimal Numbers Test



```python
# Test both Bhojpuri and English digits for decimals
decimal_tests = [
    "१२.३४",      # Bhojpuri digits
    "१२३.४५६",    # Bhojpuri digits
    "०.५",         # Bhojpuri digits
    "-१२.३४",     # Bhojpuri digits
    "१२.३४५६",    # Bhojpuri digits
    "12.34",        # English digits
    "123.456",      # English digits
    "0.5",          # English digits
    "-12.34",       # English digits
    "12.3456",      # English digits
]

print("=" * 60)
print("DECIMAL NUMBERS TEST (BHOJPURI & ENGLISH DIGITS)")
print("=" * 60)
for test in decimal_tests:
    result = normalizer_bho.normalize(test)
    print(f"Input:  {test:15} -> Output: {result}")
```

## 3. Fractions Test



```python
# Test both Bhojpuri and English digits for fractions
fraction_tests = [
    "३/४",         # Bhojpuri digits
    "१/२",         # Bhojpuri digits
    "१/४",         # Bhojpuri digits
    "१२ ३/४",      # Bhojpuri digits
    "-१/२",        # Bhojpuri digits
    "3/4",          # English digits
    "1/2",          # English digits
    "1/4",          # English digits
    "12 3/4",       # English digits
    "-1/2",         # English digits
]

print("=" * 60)
print("FRACTIONS TEST (BHOJPURI & ENGLISH DIGITS)")
print("=" * 60)
for test in fraction_tests:
    result = normalizer_bho.normalize(test)
    print(f"Input:  {test:15} -> Output: {result}")
```

## 4. Dates Test



```python
# Test both Bhojpuri and English digits for dates
date_tests = [
    "०१-०४-२०२४",      # Bhojpuri digits
    "१५-०६-२०२४",      # Bhojpuri digits
    "२०२४-०१-१५",      # Bhojpuri digits
    "१५/०६/२०२४",      # Bhojpuri digits
    "०४-०१-२०२४",      # MM-DD format, Bhojpuri digits
    "01-04-2024",        # English digits
    "15-06-2024",        # English digits
    "2024-01-15",        # English digits
    "15/06/2024",        # English digits
    "04-01-2024",        # MM-DD format, English digits
]

print("=" * 60)
print("DATES TEST (BHOJPURI & ENGLISH DIGITS)")
print("=" * 60)
for test in date_tests:
    result = normalizer_bho.normalize(test)
    print(f"Input:  {test:15} -> Output: {result}")
```

## 5. Time Test

## 11. Mixed Content Test



```python
mixed_tests = [
    "आज १५-०६-२०२४ के दिन १२:३० बजे मीटिंग है।",
    "₹१००० और ₹५०० मिलाकर ₹१५०० होता है।",
    "१२३ किलोग्राम वजन और ५० किलोमीटर दूरी।",
    "१वाँ स्थान और २वाँ स्थान।",
]

print("=" * 60)
print("MIXED CONTENT TEST")
print("=" * 60)
for test in mixed_tests:
    result = normalizer_bho.normalize(test)
    print(f"Input:  {test}")
    print(f"Output: {result}")
    print("-" * 60)
```

## 12. Batch Testing



```python
# Test multiple inputs at once
batch_tests = [
    "123",
    "१२.३४",
    "१२:३०",
    "₹१००",
    "१५-०६-२०२४",
]

print("=" * 60)
print("BATCH TESTING")
print("=" * 60)
results = normalizer_bho.normalize_list(batch_tests)
for input_text, output_text in zip(batch_tests, results):
    print(f"Input:  {input_text:15} -> Output: {output_text}")
```
