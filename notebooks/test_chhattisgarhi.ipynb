# Chhattisgarhi Text Normalization Test Cases

This notebook tests the Chhattisgarhi text normalization implementation.

```python
import sys
import os

# Add the NeMo-text-processing directory to the path
sys.path.insert(0, os.path.abspath('.'))

from nemo_text_processing.text_normalization.normalize import Normalizer

print("Imports successful!")
```

```python
# Initialize Chhattisgarhi normalizer
normalizer_cg = Normalizer(
    input_case='cased',
    lang='cg',
    cache_dir=None,  # Set to a directory path if you want to cache .far files
    overwrite_cache=False,
    post_process=True
)

print("Chhattisgarhi Text Normalizer initialized successfully!")
print(f"Language: {normalizer_cg.lang}")
```

## 1. Cardinal Numbers Test


```python
cardinal_tests = [
    "155",
    "१२३४",
    "१२३४५",
    "१२३४५६",
    "१२३४५६७",
    "१२३४५६७८",
    "-१२३",
    "-120",
    "१०००",
    "१258615",  # 1 lakh
    "१०००००००",  # 1 crore
]

print("=" * 60)
print("CARDINAL NUMBERS TEST")
print("=" * 60)
for test in cardinal_tests:
    result = normalizer_cg.normalize(test)
    print(f"Input:  {test:15} -> Output: {result}")
```

## 2. Decimal Numbers Test


```python
# Test both Chhattisgarhi and English digits for decimals
decimal_tests = [
    "१२.३४",      # Chhattisgarhi digits
    "१२३.४५६",    # Chhattisgarhi digits
    "०.५",         # Chhattisgarhi digits
    "-१२.३४",     # Chhattisgarhi digits
    "१२.३४५६",    # Chhattisgarhi digits
    "12.34",        # English digits
    "123.456",      # English digits
    "0.5",          # English digits
    "-12.34",       # English digits
    "12.3456",      # English digits
]

print("=" * 60)
print("DECIMAL NUMBERS TEST (CHHATTISGARHI & ENGLISH DIGITS)")
print("=" * 60)
for test in decimal_tests:
    result = normalizer_cg.normalize(test)
    print(f"Input:  {test:15} -> Output: {result}")
```

## 3. Fractions Test


```python
# Test both Chhattisgarhi and English digits for fractions
fraction_tests = [
    "३/४",         # Chhattisgarhi digits
    "१/२",         # Chhattisgarhi digits
    "१/४",         # Chhattisgarhi digits
    "१२ ३/४",      # Chhattisgarhi digits
    "-१/२",        # Chhattisgarhi digits
    "3/4",          # English digits
    "1/2",          # English digits
    "1/4",          # English digits
    "12 3/4",       # English digits
    "-1/2",         # English digits
]

print("=" * 60)
print("FRACTIONS TEST (CHHATTISGARHI & ENGLISH DIGITS)")
print("=" * 60)
for test in fraction_tests:
    result = normalizer_cg.normalize(test)
    print(f"Input:  {test:15} -> Output: {result}")
```

## 4. Dates Test

## 11. Mixed Content Test


```python
mixed_tests = [
    "आज १५-०६-२०२४ के दिन १२:३० बजे मीटिंग है।",
    "₹१००० और ₹५०० मिलाकर ₹१५०० होता है।",
    "१२३ किलोग्राम वजन और ५० किलोमीटर दूरी।",
    "१वाँ स्थान और २वाँ स्थान।",
]

print("=" * 60)
print("MIXED CONTENT TEST")
print("=" * 60)
for test in mixed_tests:
    result = normalizer_cg.normalize(test)
    print(f"Input:  {test}")
    print(f"Output: {result}")
    print("-" * 60)
```

## 12. Batch Testing


```python
# Test multiple inputs at once
batch_tests = [
    "123",
    "१२.३४",
    "१२:३०",
    "₹१००",
    "१५-०६-२०२४",
]

print("=" * 60)
print("BATCH TESTING")
print("=" * 60)
results = normalizer_cg.normalize_list(batch_tests)
for input_text, output_text in zip(batch_tests, results):
    print(f"Input:  {input_text:15} -> Output: {output_text}")
```
